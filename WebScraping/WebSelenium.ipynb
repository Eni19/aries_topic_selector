# SITE G1

```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup
from time import sleep
import pandas as pd

# Configurações do ChromeDriver
chrome_options = Options()
chrome_options.add_argument("--headless")  # Ativa o modo headless
chrome_options.add_argument("--disable-gpu")
chrome_options.add_argument("--no-sandbox")

driver = webdriver.Chrome(options=chrome_options)
driver.get("https://g1.globo.com/busca/?q=super+bactéria&order=relevant&species=notícias&from=2000-01-25T00%3A00%3A00-0300&to=2024-01-25T23%3A59%3A59-0300&page=1")

sleep(8)

i = 0
while i < 39: #39
    try:
        # Localiza o botão "Veja mais"
        botao = driver.find_element(By.CSS_SELECTOR, "a.fundo-cor-produto.pagination__load-more")
        botao.click()  # Clica no botão
        i += 1
        print(f"Clicando no botão 'Veja mais'... ({i})")
        sleep(5)  # Aguarda o carregamento das novas notícias
    except Exception as e:
        print("Não há mais o botão 'Veja mais' ou ocorreu um erro:", e)
        break

print(f"Total de vezes que o botão foi clicado: {i}")

sleep(5)
conteudo = BeautifulSoup(driver.page_source, 'html.parser')

news_items = []
lista = []
divs = conteudo.find_all("div", class_="widget--info__text-container")
print("Quantidade:",len(divs))

for item in divs:
    link_tag = item.find("a")
    title_tag = item.find("div", class_="widget--info__title product-color")
    if title_tag and link_tag:
        title = title_tag.get_text(strip=True)
        raw_link = link_tag.get('href')
        full_link = f"https:{raw_link}"  
        
        news_items.append({"title": title, "link": full_link})
        print(f"Clicando no link da notícia: {title}")
        print(f"Acessando link: {full_link}")
        
        # Abre o link em uma nova aba
        original_window = driver.current_window_handle
        driver.execute_script("window.open(arguments[0]);", full_link)
        sleep(5)
        
        driver.switch_to.window(driver.window_handles[-1])
        sleep(5)
        
        page_conteudo = BeautifulSoup(driver.page_source, 'html.parser')
        
       

        # Coleta informações adicionais
        headline = page_conteudo.find("h1", class_="entry-title")
        
        paragrafos = page_conteudo.find_all("p")

        texto = ""

        for paragrafo in paragrafos:
            texto += paragrafo.get_text(strip=True)
            texto += " " 

        titulo = headline.get_text(strip=True) if headline else "Título não encontrado"
        lista.append([titulo,texto,full_link])
        
        if headline:
            print(f"Manchete: {headline.get_text(strip=True)}")
        
        # Fecha a aba atual e retorna para a aba original
        driver.close()
        driver.switch_to.window(original_window)
        sleep(5)

dataset = pd.DataFrame(lista,columns=["Titulo","Corpo","Link"])
driver.quit()

dataset.to_csv("dadosG1_3.csv",sep=";")
```

# SITE GOV

```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup
import pandas as pd
import time

# Configurações do ChromeDriver
chrome_options = Options()
chrome_options.add_argument("--headless")
chrome_options.add_argument("--disable-gpu")
chrome_options.add_argument("--no-sandbox")

driver = webdriver.Chrome(options=chrome_options)

# URL de busca
palavra_chave = "amoxicilina"
url_base = f"https://www.gov.br/saude/pt-br/search?SearchableText={palavra_chave}"
driver.get(url_base)
time.sleep(3)


try:
    botao_cookie = driver.find_element(By.CSS_SELECTOR, "button.br-button.btn-manage.manage")
    botao_cookie.click()
    print("Botão de cookies clicado.")
    time.sleep(2)
except NoSuchElementException:
    print("Nenhuma barra de cookies encontrada.")

resultados = []

while True:
    # Rola a página para carregar todos os elementos
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(2)

    # Pega conteúdo da página
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    noticias = soup.find_all("li", class_="item-collective.nitf.content")
    print(f"Encontradas {len(noticias)} notícias nesta página.")

    # Se não houver notícias, pode ser um sinal de que a paginação falhou ou acabou.
    if not noticias:
        print("Nenhuma notícia encontrada nesta página. Possível fim da paginação ou erro.")
        break

    for item in noticias:
        titulo_tag = item.find("span", class_="titulo")
        a_tag = titulo_tag.find("a") if titulo_tag else None
        titulo = a_tag.get_text(strip=True) if a_tag else ""
        link = a_tag['href'] if a_tag else ""
        if link.startswith('/'):
            link = "https://www.gov.br" + link

        desc_tag = item.find("span", class_="descricao")
        descricao = desc_tag.get_text(strip=True) if desc_tag else ""

        # --- NOVA PARTE: Entrar no link e coletar o conteúdo ---
        conteudo_materia = ""
        original_window = driver.current_window_handle # Guarda a janela atual
        try:
            driver.execute_script("window.open(arguments[0]);", link) # Abre o link em uma nova aba
            time.sleep(1) # Pequena espera para a nova aba abrir
            driver.switch_to.window(driver.window_handles[-1]) # Muda para a nova aba
            time.sleep(2)  # Espera para a página carregar

            materia_soup = BeautifulSoup(driver.page_source, 'html.parser')
            content_div = materia_soup.find("div", id="content-core")
            if content_div:
                paragraphs = content_div.find_all("p")
                conteudo_materia = "\n".join([p.get_text(strip=True) for p in paragraphs])
            else:
                print(f"Div #content-core não encontrada para o link: {link}")
        except Exception as e:
            print(f"Erro ao acessar ou extrair conteúdo do link {link}: {e}")
        finally:
            driver.close() # Fecha a aba da matéria
            driver.switch_to.window(original_window) # Volta para a aba original (da busca)
            time.sleep(1) # Pequena espera para o foco voltar
            # Não é necessário driver.get(url_base) ou driver.back() aqui se usarmos abas.

        print(f"Coletado: {titulo} | {descricao} | {link} | Conteúdo (trecho): {conteudo_materia[:100]}...") # Exibe um trecho do conteúdo

        resultados.append([titulo, descricao, link, conteudo_materia])

    # Tenta encontrar e clicar no botão "Próximo"
    try:
        botao_proximo = driver.find_element(By.XPATH, "//ul[@class='paginacao listingBar']//a[@class='proximo' and contains(text(), 'Próximo »')]")
        driver.execute_script("arguments[0].scrollIntoView(true);", botao_proximo)
        time.sleep(1)
        driver.execute_script("arguments[0].click();", botao_proximo)
        print("Indo para a próxima página...")
        time.sleep(3)
    except NoSuchElementException:
        print("Não há mais páginas. Coleta finalizada.")
        break

# Fecha navegador
driver.quit()

# Salva no Excel
df = pd.DataFrame(resultados, columns=["Titulo", "Descricao", "Link", "Conteudo_Materia"])
df.to_excel("noticias_govbr.xlsx", index=False)
print("Dados salvos em 'noticias_govbr.xlsx'.")
```
